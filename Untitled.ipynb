```python

# Import library yang dibutuhkan
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
import matplotlib.pyplot as plt

# 1. Memuat dataset (Data kelulusan mahasiswa)
# Contoh data: 'status' adalah target (lulus atau tidak)
data = {
    'nilai_tugas': [80, 85, 78, 92, 70, 65, 90, 75, 60, 88],
    'nilai_uts': [75, 80, 85, 90, 60, 55, 88, 65, 50, 85],
    'nilai_uas': [78, 82, 84, 93, 68, 60, 87, 72, 55, 89],
    'status': ['Lulus', 'Lulus', 'Lulus', 'Lulus', 'Tidak Lulus',
               'Tidak Lulus', 'Lulus', 'Tidak Lulus', 'Tidak Lulus', 'Lulus']
}

df = pd.DataFrame(data)

# 2. Memproses data
# Mengubah 'status' menjadi nilai numerik (0 untuk 'Tidak Lulus', 1 untuk 'Lulus')
df['status'] = df['status'].map({'Tidak Lulus': 0, 'Lulus': 1})

# Memisahkan fitur (X) dan target (y)
X = df.drop(columns='status')
y = df['status']

# 3. Membagi data ke dalam data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 4. Melatih model Decision Tree
clf = DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)

# 5. Memprediksi hasil pada data uji
y_pred = clf.predict(X_test)

# 6. Mengukur akurasi model
print("Akurasi Model:", accuracy_score(y_test, y_pred))
print("\nLaporan Klasifikasi:\n", classification_report(y_test, y_pred))

# Visualisasi pohon keputusan
plt.figure(figsize=(12,8))
tree.plot_tree(clf, filled=True, feature_names=X.columns, class_names=['Tidak Lulus', 'Lulus'])
plt.show()
```
